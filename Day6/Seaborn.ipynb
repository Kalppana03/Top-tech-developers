```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("employee_promotion.csv")
```

```python
df.head
```

```python
df.isnull().sum()
```

```python
df= df.dropna()
```

```python
df= df.drop_duplicates()
```

```python
df
```

```python
sns.scatterplot(data=df,x='previous_year_rating',y='avg_training_score',hue='is_promoted')
plt.title('previous_year_rating vs avg_training_score')
plt.show()
```

```python
sns.barplot(data=df,x="is_promoted",y="length_of_service")
sns.set_style("darkgrid")
plt.title("Average length of service for promotion")
plt.show()
```

```python
sns.histplot(data=df, x="length_of_service", kde=True, bins=5)
plt.title('Distribution service')
plt.show()
```

```python
sns.boxplot(data=df, x='recruitment_channel', y="previous_year_rating")
plt.show()
```

```python
sns.violinplot(data=df, x='recruitment_channel', y='previous_year_rating')
plt.show()
```

```python
# Compute correlation matrix
corr = df.select_dtypes(include=['number']).corr()
# Draw heatmap with annotations
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of df Features')
plt.show()
```

```python
sns.pairplot(df, hue='is_promoted')
plt.suptitle('Pairplot of df Dataset', y=1.02)
plt.show()
```
