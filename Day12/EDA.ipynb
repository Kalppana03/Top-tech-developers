```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
age_groups = ['20–30', '31–40', '41–50', '51+']
frequencies = [40, 50, 25, 5]
```

```python


data = {
    'Age_Group': ['20-30','31-40','41-50','51+'],
    'Frequency': [40,50,25,5],
    'Percentage':[33.3,41.7, 20.8, 4.2]
}
df = pd.DataFrame(data)

plt.figure(figsize=(7, 6))
ax = sns.barplot(x='Age_Group', y='Frequency', data=df)  # Change palette here

# Add percentage labels on top of each bar
for p, pct in zip(ax.patches, df['Percentage']):
    height = p.get_height()
    ax.text(
        x=p.get_x() + p.get_width() / 2,
        y=height + 0.5,
        s=f'{pct:.1f}%',
        ha='center',
        fontsize=8
    )

# Axis labels and tick font sizes
ax.set_xlabel("Age_Group", fontsize=10)
ax.set_ylabel("Frequency", fontsize=10)
ax.tick_params(axis='x', labelsize=8)
ax.tick_params(axis='y', labelsize=8)

# Add chart title
plt.title("Age Distribution of Respondents", fontsize=12, weight='bold')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'Level': ['Very Aware', 'Somewhat Aware', 'Slightly Aware', 'Not Aware'],
    'Frequency': [30, 45, 25, 20]
}
df = pd.DataFrame(data)

ax = sns.barplot(x='Level', y='Frequency', data=df)

# Add smaller text labels on top of each bar
for p in ax.patches:
    height = p.get_height()
    ax.text(
        x=p.get_x() + p.get_width() / 2, 
        y=height + 0.5,  # a little above the bar
        s=f'{int(height)}', 
        ha='center', 
        fontsize=8  # smaller font size
    )

# Custom legend with smaller font size
plt.legend(['Frequency'], loc='upper right', fontsize=8)

plt.tight_layout()
plt.show()

```
